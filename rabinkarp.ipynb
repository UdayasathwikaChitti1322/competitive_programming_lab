```python
def rabin_karp(text, pattern, q):
    d = 256
    n = len(text)
    m = len(pattern)

    p_hash = 0
    t_hash = 0
    h = 1

    spurious_hits = 0
    actual_hits = 0

    # Compute h = d^(m-1) % q
    for i in range(m - 1):
        h = (h * d) % q

    # Initial hash values
    for i in range(m):
        p_hash = (d * p_hash + ord(pattern[i])) % q
        t_hash = (d * t_hash + ord(text[i])) % q

    print("\nPattern Hash:", p_hash)
    print("--------------------------------------------------")

    # Slide pattern over text
    for i in range(n - m + 1):
        window = text[i:i+m]
        print(f"Window '{window}' at index {i}")

        if p_hash == t_hash:
            print(" Hash matched!", end=" ")

            if window == pattern:
                print("=> Match Found ✅")
                actual_hits += 1
            else:
                print("=> Spurious Hit ❌")
                spurious_hits += 1
        else:
            print(" No Match ❌")

        # Calculate next hash
        if i < n - m:
            t_hash = (d * (t_hash - ord(text[i]) * h) + ord(text[i + m])) % q

            if t_hash < 0:
                t_hash += q

        print("--------------------------------------------------")

    print("\nRESULT:")
    print("Total Spurious Hits:", spurious_hits)
    print("Actual Matches:", actual_hits)
    print("After removing spurious hits:", actual_hits)

    if actual_hits > 0:
        print("Pattern Found in Text ✅")
    else:
        print("Pattern Not Found ❌")


# 🔹 Taking user input
text = input("Enter the text: ")
pattern = input("Enter the pattern: ")
q = int(input("Enter a prime number (e.g., 101): "))

# Condition check
if len(pattern) > len(text):
    print("Pattern length should be less than or equal to text length")
else:
    rabin_karp(text, pattern, q)
```
