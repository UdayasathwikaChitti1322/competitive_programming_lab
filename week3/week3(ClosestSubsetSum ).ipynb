```python
import bisect

def subset_sums(arr):
    sums = []
    n = len(arr)
    for mask in range(1 << n):
        s = 0
        for i in range(n):
            if mask & (1 << i):
                s += arr[i]
        sums.append(s)
    return sums


T = int(input())

for _ in range(T):
    N, S = map(int, input().split())
    A = list(map(int, input().split()))

    mid = N // 2
    left = A[:mid]
    right = A[mid:]

    L = subset_sums(left)
    R = subset_sums(right)

    R.sort()

    ans = float('inf')

    for l in L:
        target = S - l
        idx = bisect.bisect_left(R, target)

        if idx < len(R):
            ans = min(ans, abs(l + R[idx] - S))
        if idx > 0:
            ans = min(ans, abs(l + R[idx - 1] - S))

    print(ans)
```
